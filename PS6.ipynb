```python
def cal_mean(df,col):
    numerator=0
    for i in df[col]:
        numerator=numerator+i
    return numerator/len(df)
```

```python
print(cal_mean(IrisSetosa,"SepalLengthCm"))
```

```python
def cal_median(df):
    list=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
    for i in list:
        n=len(df[i])
        index=n//2
        if n%2==0:
            print(sorted(df[i])[index])
        else:
            print(sum(sorted(df[i]))[index-1:index+1]/2)
    
```

```python
cal_median(IrisSetosa)
```

```python
IrisSetosa["SepalWidthCm"].median()
```

```python
import math
def findStd(data):
    list=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
    for j in list:
        n = len(data[j])
        m=cal_mean(data,j)
        numerator=0
        for i in (data[j]):
            numerator+=pow(i-m,2)
        print("Variance of ",j)
        print(numerator/(n-1))
        std=math.sqrt(numerator/(n-1))
        print("Standard Deviation of ",j)
        print(std)
        
```

```python
findStd(IrisSetosa)
```

```python
def findRange(data):
    list=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']
    for j in list:
        n = len(data[j])
        maxm=-9999999
        minm=9999999
        for i in data[j]:
            if i > maxm:
                maxm=i
            if i<minm:
                minm=i
        rangeValue=maxm-minm
        print("Range value of ",j)
        print(rangeValue)
```

```python
findRange(IrisSetosa)
```
